```python
import matplotlib.pyplot as plt
import soundfile as sf
from scipy.signal import lfilter, freqz, butter, filtfilt
from scipy.fft import fft, rfft
import numpy as np
from settings import *
```

```python
for number in NUMBERS:
    for person in PEOPLE:
        audio_path = FINAL_FOLDER + person + "_" + number + "_final" + AUDIO_FORMAT
        num_audio, sample_rate = sf.read(audio_path)

        f_magnitude = np.abs(rfft(num_audio))
        # f_magnitude = 20 * np.log10(np.abs(rfft(num_audio)))

        # Plot to compare the signal and frequency using subplot
        fig, (ax1, ax2) = plt.subplots(2, 1)
        fig.suptitle(number + " " + person)
        ax1.plot(num_audio)
        ax1.set_ylabel('Amplitude')
        ax2.plot(f_magnitude[0:4000])
        ax2.set_ylabel('Frequency')
        ax2.set_xlabel('Sample')
        plt.show()
```
